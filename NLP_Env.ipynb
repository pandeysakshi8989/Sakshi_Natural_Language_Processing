# NLP Overview 

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import pos_tag
```

```python
# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
# Input sentence
text = "Natural Language Processing is very interesting and useful"

print("Original Text:")
print(text)
```

```python
# 1. Tokenization
tokens = word_tokenize(text)
print("\nTokens:")
print(tokens)
```

```python
# 2. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered = [word for word in tokens if word.lower() not in stop_words]
print("\nAfter Stopword Removal:")
print(filtered)
```

```python
# 3. Stemming
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in filtered]
print("\nAfter Stemming:")
print(stemmed)
```

```python
# 4. POS Tagging
pos = pos_tag(tokens)
print("\nPart of Speech Tagging:")
print(pos)
```

# NLP Pipeline

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
text = "Artificial Intelligence is changing the world. Google is located in California. Elon Musk is the CEO of Tesla."
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print("Segmentation:", sentences)
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print("Tokenization:", tokens)
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed = [ps.stem(w) for w in tokens]
print("Stemming:", stemmed)
```

```python
from nltk.stem import WordNetLemmatizer

lemm = WordNetLemmatizer()
lemmatized = [lemm.lemmatize(w) for w in tokens]
print("Lemmatization:", lemmatized)
```

```python
pos_tags = nltk.pos_tag(tokens)
print("POS Tagging:", pos_tags)
```

```python
named_entities = nltk.ne_chunk(pos_tags)
print("Named Entities:", named_entities)
```

# NLP LIBRARIES

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

text = "I love Natural Language Processing!"
tokens = word_tokenize(text)
print(tokens)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from gensim.models import Word2Vec

sentences = [["I", "love", "NLP"], ["Word embeddings are fun"]]
model = Word2Vec(sentences, vector_size=50, window=2, min_count=1, workers=4)
print(model.wv['NLP'])
```

```python
from textblob import TextBlob

text = TextBlob("I love Natural Language Processing!")
print(text.sentiment)
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
result = classifier("I love studying Natural Language Processing!")
print(result)
```
