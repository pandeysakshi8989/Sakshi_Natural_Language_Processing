```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree
```

```python
# Download required packages (run once)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
# Sample text
text = "Apple Inc. was founded by Steve Jobs in Cupertino on April 1, 1976."
```

```python
# Tokenization
tokens = word_tokenize(text)
```

```python
# POS Tagging
pos_tags = pos_tag(tokens)
```

```python
# Named Entity Recognition
ner_tree = ne_chunk(pos_tags)
```

```python
# Display Named Entities
print("Named Entities:\n")
for subtree in ner_tree:
    if isinstance(subtree, Tree):
        entity = " ".join([word for word, tag in subtree.leaves()])
        label = subtree.label()
        print(entity, "->", label)
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

from nltk import sent_tokenize, word_tokenize, pos_tag, ne_chunk

# Sample text
text = "Sundar Pichai is the CEO of Google. He was born in Chennai on July 12, 1972."

# Sentence Segmentation
sentences = sent_tokenize(text)

for sent in sentences:
    tokens = word_tokenize(sent)
    pos_tags = pos_tag(tokens)
    named_entities = ne_chunk(pos_tags)

    print("\nNamed Entities in Sentence:")
    print(named_entities)
```

```python
# Sample reviews
reviews = [
    "I love this product, it is amazing!",
    "This is the worst purchase I ever made.",
    "The product is okay, not great."
]
```

```python
# Sentiment Analysis
for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(f"Review: {review}\nSentiment: {sentiment}\nPolarity: {polarity}\n")
```

## Practical 10 – Rule-Based Chatbot

Aim:
To design a simple rule-based chatbot using pattern matching to answer user queries.

Procedure

Import required libraries re and random.

Create a dictionary of user patterns and responses.

Preprocess user input (convert to lowercase).

Match input using regular expressions.

Display corresponding response.

Provide default reply if no match found.

Loop chatbot for continuous interaction.

Test chatbot with various inputs.

Save the script for future use.

```python
import re
import random
```

```python
# Predefined patterns and responses
responses = {
    "hello": ["Hi there!", "Hello! How can I help you?"],
    "how are you": ["I'm a bot, I'm always fine!", "Doing great! How about you?"],
    "bye": ["Goodbye!", "See you later!"]
}
```

```python
# Chatbot loop
while True:
    user_input = input("You: ").lower()
    if user_input == "exit":
        print("Bot: Goodbye!")
        break
    response_found = False
    for pattern, resp_list in responses.items():
        if re.search(pattern, user_input):
            print("Bot:", random.choice(resp_list))
            response_found = True
            break
    if not response_found:
        print("Bot: I am sorry, I don't understand.")
```

Result: Program executed successfully

```python
import re
import random

patterns = {
    r'hi|hello|hey': ["Hello!", "Hi there!", "Hey! How can I help you?"],
    r'how are you': ["I am fine!", "Doing great!", "I am here to help you."],
    r'what is your name': ["I am an NLP Chatbot.", "You can call me NLP Bot."],
    r'bye|exit': ["Goodbye!", "See you soon!", "Have a great day!"]
}

def chatbot_response(user_input):
    user_input = user_input.lower()
    for pattern, responses in patterns.items():
        if re.search(pattern, user_input):
            return random.choice(responses)
    return "Sorry, I did not understand that."

print("Chatbot is running! Type 'bye' to exit.")

while True:
    user = input("You: ")
    response = chatbot_response(user)
    print("Bot:", response)
    if re.search(r'bye|exit', user.lower()):
        break
```

```python
# BFS
def bfs(start):
    visited = []
    queue = deque([start])
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])
    return visited

# DFS
def dfs(start, visited=None):
    if visited is None:
        visited = []
    visited.append(start)
    for node in graph[start]:
        if node not in visited:
            dfs(node, visited)
    return visited
```

## PRACTICAL – 13 (i)
Best-First Search (Heuristic-Based Search)

Aim

To implement Best-First Search algorithm using heuristic values for AI pathfinding.

Procedure

Represent the problem as a graph with nodes and edges.

Assign heuristic values (estimated cost to goal) to each node.

Initialize a priority queue with the start node.

Pop the node with the lowest heuristic from the queue.

Mark it as visited and add its neighbors to the queue based on heuristic.

Repeat until the goal node is reached.

Track visited nodes to avoid cycles.

Display the path and heuristic cost.

Save code for reference.

```python
import heapq

# Graph representation
graph = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Heuristic values (estimated cost to goal F)
heuristic = {'A': 5, 'B': 4, 'C': 2, 'D': 3, 'E': 1, 'F': 0}

def best_first_search(start, goal):
    visited = set()
    queue = []
    heapq.heappush(queue, (heuristic[start], start))
    path = []

    while queue:
        h, node = heapq.heappop(queue)
        if node not in visited:
            path.append(node)
            visited.add(node)
            if node == goal:
                break
            for neighbor in graph[node]:
                if neighbor not in visited:
                    heapq.heappush(queue, (heuristic[neighbor], neighbor))
    return path

print("Best-First Search Path:", best_first_search('A','F'))
```

## PRACTICAL – 13 (ii) A* Search Algorithm

Aim

To implement the A* Search algorithm using both actual cost and heuristic values for AI pathfinding.

Procedure

Represent the problem as a graph with edges and costs.

Assign heuristic values to each node.

Initialize a priority queue with (f = g + h, node) where g = actual cost, h = heuristic.

Pop the node with the lowest f value.

Expand neighbors and calculate their f = g + h.

Track visited nodes to avoid cycles.

Repeat until the goal is reached.

Display the optimal path and total cost.

Compare with Best-First Search.

```python
import heapq

# Graph with edge costs
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 3, 'E': 1},
    'C': {'F': 5},
    'D': {},
    'E': {'F': 2},
    'F': {}
}

# Heuristic values
heuristic = {'A': 5, 'B': 4, 'C': 2, 'D': 3, 'E': 1, 'F': 0}

def a_star_search(start, goal):
    open_set = []
    heapq.heappush(open_set, (heuristic[start], 0, start, [start]))  # (f, g, node, path)
    visited = set()

    while open_set:
        f, g, node, path = heapq.heappop(open_set)
        if node == goal:
            return path, g
        if node not in visited:
            visited.add(node)
            for neighbor, cost in graph[node].items():
                if neighbor not in visited:
                    g_new = g + cost
                    f_new = g_new + heuristic[neighbor]
                    heapq.heappush(open_set, (f_new, g_new, neighbor, path + [neighbor]))
    return None

path, cost = a_star_search('A', 'F')
print("A* Search Path:", path)
print("Total Cost:", cost)
```

## Practical 15 (i) - Rule-Based Chatbot

Aim

To design a simple chatbot that answers user queries using predefined patterns.

Procedure

Import required libraries: re and random.

Create a dictionary of possible user queries and responses.

Preprocess user input: lowercase, remove punctuation.

Match input with predefined patterns using regex.

Return the corresponding response; if no match, return a default reply.

Loop the chatbot for multiple queries.

Test the chatbot with different inputs.

```python
import re
import random

patterns = {
    r'hi|hello|hey': ["Hello!", "Hi there!", "Hey! How can I help you?"],
    r'how are you': ["I am fine!", "Doing great!", "I am here to help you."],
    r'what is your name': ["I am an NLP Chatbot.", "You can call me NLP Bot."],
    r'bye|exit': ["Goodbye!", "See you soon!", "Have a great day!"]
}

def chatbot_response(user_input):
    user_input = user_input.lower()
    for pattern, responses in patterns.items():
        if re.search(pattern, user_input):
            return random.choice(responses)
    return "Sorry, I did not understand that."

print("Chatbot is running! Type 'bye' to exit.")
while True:
    user = input("You: ")
    response = chatbot_response(user)
    print("Bot:", response)
    if re.search(r'bye|exit', user.lower()):
        break
```

## Practical 15 (ii) - Text Summarizer

Aim

To generate a concise summary of a large text using NLP techniques.

Procedure

Import nltk and heapq.

Load or input a text document.

Tokenize text into sentences and words.

Compute word frequencies, ignoring stopwords.

Score each sentence based on important words.

Select top sentences to form the summary.

Display summarized text.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import heapq

nltk.download('punkt')
nltk.download('stopwords')

text = """
Artificial Intelligence (AI) is a branch of computer science that aims to create intelligent machines.
It has applications in various fields such as healthcare, finance, education, and autonomous vehicles.
Machine Learning, a subset of AI, allows systems to learn from data without explicit programming.
Natural Language Processing enables machines to understand and interpret human language.
AI continues to grow rapidly and impacts everyday life in multiple ways.
"""

stop_words = set(stopwords.words("english"))
words = word_tokenize(text.lower())
word_frequencies = {}

for word in words:
    if word not in stop_words and word.isalpha():
        word_frequencies[word] = word_frequencies.get(word, 0) + 1

sentences = sent_tokenize(text)
sentence_scores = {}
for sent in sentences:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies:
            sentence_scores[sent] = sentence_scores.get(sent, 0) + word_frequencies[word]

summary_sentences = heapq.nlargest(3, sentence_scores, key=sentence_scores.get)
summary = " ".join(summary_sentences)
print("Summary:\n", summary)
```

## **Practical 13 – A* / Best-First Search**

**Q1. What is heuristic search?**
Search that uses estimated cost to goal to guide exploration.

**Q2. Define Best-First Search.**
Expands nodes with lowest heuristic value first.

**Q3. How does A* algorithm work?**
Combines actual cost (g) and heuristic (h) to select optimal path.

**Q4. Difference between Best-First and A* Search.**
Best-First uses only heuristic; A* uses g + h for optimality.

**Q5. Role of heuristic values?**
Estimate distance or cost to reach goal.

**Q6. Data structure used in Best-First Search?**
Priority queue.

**Q7. What is f(n) in A* Search?**
f(n) = g(n) + h(n), total estimated cost of node.

**Q8. Real-world application of A* Search?**
Pathfinding in games and GPS systems.

**Q9. How does A* ensure optimal path?**
By considering both actual cost and heuristic (admissible h).

**Q10. How do you avoid cycles in heuristic search?**
Maintain a visited set to skip already explored nodes.
